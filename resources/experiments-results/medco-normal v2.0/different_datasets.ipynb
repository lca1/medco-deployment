```python
#DIFFERENT DATASETS

file = open("./1000/queryA/queryA_1000.txt","r") 

# QUERY A (#1000) Overall
print "QUERY A (#1000): OVERALL\n"

# overall
split_tag = "i2b2 query&quot;:"

with file as f:
    for line in f:
        cnt = ''.join(line).split(split_tag)
        if len(cnt)>1:
            print ''.join(cnt[1]).split(",")[0]
            
file.close()
```
